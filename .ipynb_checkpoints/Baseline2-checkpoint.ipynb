```python
import gzip
import tensorflow as tf
import array
import random
from tensorflow.keras import Model
from collections import defaultdict
import random

def readJSON(path):
  for l in gzip.open(path, 'rt', encoding="utf8"):
    d = eval(l)
    yield d
    
def readJSON2(path):
  for l in gzip.open(path, 'rt', encoding="utf8"):
    d = eval(l)
    u = str(d['user_id'])
    try:
     b = str(d['book_id'])
    except Exception as e:
     b = None
    yield u,b, d
```

```python
userIDs = {}
itemIDs = {}
trainSet = []
validationSet = []
testSet = []
interactions = []
bookCount = defaultdict(int)
totalRead = 0
iterations = 0 
for u,b,d in readJSON2("goodreads_reviews_dedup.json.gz"):
    if iterations == 100000: break
    # Get first 100000 reviews to later split into train/data/test set
    interactions.append((u,b,d))
    iterations += 1    
```

```python
# Shuffle dataset and split
random.shuffle(interactions)
trainSet = interactions[0:80000]
validation = interactions[80000:90000]
testSet = interactions[90000:100000]
#Count number of interactions for each book
usersPerItem = defaultdict(set) # U_i from class slides
itemsPerUser = defaultdict(set) # I_u from class slides
for user,book,d in trainSet:
    totalRead += 1
    bookCount[book] += 1
    usersPerItem[book].add(user)
    itemsPerUser[user].add(book)
items = list(usersPerItem.keys())
```

```python
mostPopular = [(bookCount[x], x) for x in bookCount]
mostPopular.sort()
mostPopular.reverse()
returnPopular = set()
count = 0
#Extract books that account for 40% of all interactions
for ic, i in mostPopular:
  count += ic
  returnPopular.add(i)
  if count > totalRead*0.28: break
```

```python
# Create negative samples for validation set.
validationSet = []
for user,book,d in validation:
    validationSet.append((user,book,1))
    bookNeg = random.choice(items) # negative sample
    while bookNeg in itemsPerUser[user]:
        bookNeg = random.choice(items)
    validationSet.append((user,bookNeg,0))  
```

```python
predictions_lst = []
count = 0
labels = []
for sample in validationSet:
    user = sample[0]
    book = sample[1]
    if book in returnPopular:
        predictions_lst.append(1)
    else:
        predictions_lst.append(0)
    labels.append(sample[2])
```

```python
#accuracy
count = 0
for i in range(len(predictions_lst)):
    if predictions_lst[i] == labels[i]:
        count += 1
accuracy = count/len(predictions_lst)
accuracy
```
